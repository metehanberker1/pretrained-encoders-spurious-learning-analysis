```python
import pandas as pd
from difflib import SequenceMatcher
import re
from nltk.corpus import stopwords

# Load stop words
stop_words = set(stopwords.words("english"))

def create_augmented_dataset(file_path, lower_bound=15, upper_bound=20):
    """
    Create an augmented dataset with detailed fields for counterfactual words, ratios, and changes.
    Rows are included only if both sentences' ratio percentages are within the specified range.

    Args:
        file_path (str): Path to the input TSV file.
        lower_bound (float): Lower bound for the span percentage (inclusive).
        upper_bound (float): Upper bound for the span percentage (inclusive).

    Returns:
        pd.DataFrame: An augmented dataset with detailed fields.
    """
    # Load the data
    data = pd.read_csv(file_path, sep="\t")

    # Map sentiment values to binary
    data['Sentiment'] = data['Sentiment'].map({"Positive": 1, "Negative": 0})

    # Group by batch_id
    grouped = data.groupby("batch_id")
    augmented_rows = []

    for batch_id, group in grouped:
        if len(group) != 2:
            continue  # Skip if the pair is incomplete

        # Extract sentences and sentiments
        first_sentence = group.iloc[0]["Text"]
        second_sentence = group.iloc[1]["Text"]
        first_sentence_sentiment = group.iloc[0]["Sentiment"]
        second_sentence_sentiment = group.iloc[1]["Sentiment"]

        # Tokenize sentences into words while keeping punctuation attached
        def tokenize_with_punctuation(sentence):
            return re.findall(r'\w+|[^\w\s]', sentence)

        first_tokens = tokenize_with_punctuation(first_sentence)
        second_tokens = tokenize_with_punctuation(second_sentence)

        # Detect changes using SequenceMatcher
        matcher = SequenceMatcher(None, first_tokens, second_tokens)
        first_sentence_changes = []
        second_sentence_changes = []
        first_counterfactual_words = []
        second_counterfactual_words = []

        temp_span_first = []
        temp_span_second = []
        for tag, i1, i2, j1, j2 in matcher.get_opcodes():
            if tag == "equal":
                # Append spans if temporary spans exist
                if temp_span_first:
                    clean_span = clean_token_span(temp_span_first)
                    if clean_span:
                        first_counterfactual_words.append(clean_span)
                    temp_span_first = []
                if temp_span_second:
                    clean_span = clean_token_span(temp_span_second)
                    if clean_span:
                        second_counterfactual_words.append(clean_span)
                    temp_span_second = []
                # Add unchanged parts
                unchanged_text_first = " ".join(first_tokens[i1:i2])
                unchanged_text_second = " ".join(second_tokens[j1:j2])
                first_sentence_changes.append(unchanged_text_first)
                second_sentence_changes.append(unchanged_text_second)
            elif tag in {"replace", "insert", "delete"}:
                # Append changed spans
                changed_first = first_tokens[i1:i2]
                changed_second = second_tokens[j1:j2]
                if changed_first:
                    temp_span_first.extend(changed_first)
                    first_sentence_changes.append(f"{{{' '.join(changed_first)}}}")
                if changed_second:
                    temp_span_second.extend(changed_second)
                    second_sentence_changes.append(f"{{{' '.join(changed_second)}}}")

        # Append the last spans if any
        if temp_span_first:
            clean_span = clean_token_span(temp_span_first)
            if clean_span:
                first_counterfactual_words.append(clean_span)
        if temp_span_second:
            clean_span = clean_token_span(temp_span_second)
            if clean_span:
                second_counterfactual_words.append(clean_span)

        # Calculate word counts and ratios
        first_word_count = len(first_tokens)
        second_word_count = len(second_tokens)
        first_ratio = (len(" ".join(first_counterfactual_words).split()) / first_word_count) * 100 if first_word_count > 0 else 0
        second_ratio = (len(" ".join(second_counterfactual_words).split()) / second_word_count) * 100 if second_word_count > 0 else 0

        # Filter by the ratio bounds for both sentences
        if not (lower_bound <= first_ratio <= upper_bound and lower_bound <= second_ratio <= upper_bound):
            continue

        # Join changes into strings
        first_sentence_changes_str = " ".join(first_sentence_changes)
        second_sentence_changes_str = " ".join(second_sentence_changes)

        # Add row to augmented dataset
        augmented_rows.append({
            "batch_id": batch_id,
            "first_sentence": first_sentence,
            "second_sentence": second_sentence,
            "first_sentence_changes": first_sentence_changes_str,
            "second_sentence_changes": second_sentence_changes_str,
            "first_sentence_counterfactual_words": first_counterfactual_words,
            "second_sentence_counterfactual_words": second_counterfactual_words,
            "first_sentence_ratio": first_ratio,
            "second_sentence_ratio": second_ratio,
            "first_sentence_sentiment": first_sentence_sentiment,
            "second_sentence_sentiment": second_sentence_sentiment
        })

    # Create DataFrame
    return pd.DataFrame(augmented_rows)

def clean_token_span(span):
    """
    Cleans a token span by removing leading/trailing punctuation and spaces.
    Filters out spans that are entirely stopwords or standalone punctuation.

    Args:
        span (list of str): Token span to clean.

    Returns:
        str: Cleaned span or None if it is invalid.
    """
    # Join tokens to form the span string
    span_str = " ".join(span)

    # Strip punctuation and spaces from the beginning and end
    span_str = span_str.strip().strip('.,!?;"\'')

    # Check if the span is empty or consists of only stopwords/punctuation
    if not span_str or all(word.lower() in stop_words for word in span_str.split()):
        return None

    return span_str
```

```python
train_path = "/content/counterfactually-augmented-data/sentiment/combined/paired/train_paired.tsv"
test_path = "/content/counterfactually-augmented-data/sentiment/combined/paired/test_paired.tsv"
dev_path = "/content/counterfactually-augmented-data/sentiment/combined/paired/dev_paired.tsv"

lower_bound = 0
upper_bound = 100

print("Processing train dataset...")
augmented_train = create_augmented_dataset(train_path, lower_bound=lower_bound, upper_bound=upper_bound)
augmented_train.to_csv("augmented_train.csv", index=False)

print("Processing test dataset...")
augmented_test = create_augmented_dataset(test_path, lower_bound=lower_bound, upper_bound=upper_bound)
augmented_test.to_csv("augmented_test.csv", index=False)

print("Processing dev dataset...")
augmented_dev = create_augmented_dataset(dev_path, lower_bound=lower_bound, upper_bound=upper_bound)
augmented_dev.to_csv("augmented_dev.csv", index=False)

print("Augmented datasets saved:")
print(" - augmented_train.csv")
print(" - augmented_test.csv")
print(" - augmented_dev.csv")
```

CREATING THE DATASETS

```python
import pandas as pd

def split_paired_sentences_by_split(file_paths):
    """
    Split paired sentences into original and augmented datasets for train, test, and dev splits.
    Maps sentiment values to binary (Positive -> 1, Negative -> 0).

    Args:
        file_paths (dict): Dictionary containing paths to the train, test, and dev paired datasets.

    Returns:
        dict: Dictionary containing the six datasets (original and augmented for train, test, and dev).
    """
    splits = ["train", "test", "dev"]
    datasets = {}

    for split in splits:
        # Load the paired dataset
        file_path = file_paths[split]
        data = pd.read_csv(file_path, sep="\t")

        # Split the data
        original_sentences = []
        augmented_sentences = []

        for _, row in data.iterrows():
            # First sentence: original
            original_sentences.append({
                "sentence": row["Text"],
                "sentiment": 1 if row["Sentiment"] == "Positive" else 0,
                "batch_id": row["batch_id"]
            })

            # Second sentence: augmented
            reversed_sentiment = 1 if row["Sentiment"] == "Negative" else 0
            augmented_sentences.append({
                "sentence": row["Text"],
                "sentiment": reversed_sentiment,
                "batch_id": row["batch_id"]
            })

        # Create DataFrames
        datasets[f"{split}_original"] = pd.DataFrame(original_sentences)
        datasets[f"{split}_augmented"] = pd.DataFrame(augmented_sentences)

    return datasets
```

```python
# Define file paths
file_paths = {
    "train": "/content/counterfactually-augmented-data/sentiment/combined/paired/train_paired.tsv",
    "test": "/content/counterfactually-augmented-data/sentiment/combined/paired/test_paired.tsv",
    "dev": "/content/counterfactually-augmented-data/sentiment/combined/paired/dev_paired.tsv"
}

# Split paired sentences into six datasets
datasets = split_paired_sentences_by_split(file_paths)

# Save each dataset to a CSV
for split_name, dataset in datasets.items():
    output_path = f"{split_name}_dataset.csv"
    dataset.to_csv(output_path, index=False)
    print(f"Saved {split_name} dataset to {output_path}")
```

```python
def push_dataset_to_huggingface(dataset_path, split_name, hf_username, repo_name):
    """
    Push a single dataset split to Hugging Face Hub.

    Args:
        dataset_path (str): Path to the CSV file containing the dataset.
        split_name (str): Name of the dataset split (e.g., train_original).
        hf_username (str): Hugging Face username.
        repo_name (str): Repository name on Hugging Face Hub.
    """
    # Load the dataset from CSV
    df = pd.read_csv(dataset_path)
    dataset = Dataset.from_pandas(df)

    # Push to Hugging Face
    dataset.push_to_hub(f"{hf_username}/{repo_name}", split=split_name)
    print(f"Dataset '{split_name}' pushed to Hugging Face Hub under {hf_username}/{repo_name}")
```
